```python
import pandas as pd
import latex
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import scienceplots
import seaborn as sns
```

```python
# get a table to save the data
# the column names
col_names = ['Year', 'Average Replies', 'type']
# the records
first_row = {'Year': 2020, 'Average Replies': 1.43, 'type': 'Tweet'}
second_row = {'Year': 2020, 'Average Replies': 0.61, 'type': 'Meme'}
third_row = {'Year': 2021, 'Average Replies': 0.99, 'type': 'Tweet'}
fourth_row = {'Year': 2021, 'Average Replies': 2.2, 'type': 'Meme'}
fifth_row = {'Year': 2022, 'Average Replies': 0.63, 'type': 'Tweet'}
sixth_row = {'Year': 2022, 'Average Replies': 2.57, 'type': 'Meme'}

# the table
df = pd.DataFrame([first_row, second_row, third_row, fourth_row, fifth_row, sixth_row], columns=col_names)
df
```

```python
# create the figure
plt.style.use(['science','ieee', 'no-latex'])

plt.rcParams.update({
    "font.family": "serif",
    # Use LaTeX default serif font.
    "font.serif": ["Times New Roman"],
    # Use specific cursive fonts.
    "font.cursive": ["Comic Neue", "Times New Roman"],
})

# plot the bars for meme and text posts
fig = plt.figure()
ax = fig.add_subplot(111)
# sns.barplot(x='Year', y='Average Replies', hue='type', data=df, ax=ax, palette=['#00a8e1', '#ff6600'])
sns.barplot(x='Year', y='Average Replies', hue='type', data=df, ax=ax, palette='muted')

ax.set_xlabel('Year')
ax.set_ylabel('Average Replies')
# ax.set_title('Average Replies for Meme and Text Posts')
# put the legend to the top and change the name of the legend
ax.legend(loc='upper left', fontsize=10)

# ax.legend(loc='upper right')
# plt.show()
plt.savefig('meme_plot.pdf')
```
